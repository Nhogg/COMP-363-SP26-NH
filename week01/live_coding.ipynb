```python
def merge(a, b):

    # leftmost indices for input arrays:
    i = j = 0

    # initialize output array
    c = []

    # process both arrays first
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            c.append(a[i])  # take leftmost from a and add to output
            i = i + 1  # increment leftmost index for a
        else:
            c.append(b[j])  # do the same from b
            j = j + 1  # increment leftmost index for b
    while j < len(b):  # Just in case b has unprocessed elements
        c.append(b[j])  # add the rest of b's elements to output
        j += 1  # Just in case a has unprocessed elements
    while i < len(a):  # add the rest of a's elements to output
        c.append(a[i])
        i += 1
    return c  # Done
```
