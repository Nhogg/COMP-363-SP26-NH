# 3-SAT

```python
from __future__ import annotations
import time
import random


def make_random_3sat(num_vars: int, num_clauses: int) -> list[list[int]]:
    """
    Create a random 3SAT instance.

    Each clause is a list of 3 literals.
    A positive integer k means variable x_k.
    A negative integer -k means NOT x_k.

    Example clause: [1, -3, 4] means
    (x1 OR NOT x3 OR x4)
    """
    formula: list[list[int]] = []

    for _ in range(num_clauses):
        # For each clause select three literals in the range [1,n]
        chosen_vars = random.sample(range(1, num_vars + 1), 3)
        # Initialize a list to hold a clause based on these literals
        clause: list[int] = []
        # Compose the clause by negating some literals at random
        for var in chosen_vars:
            if random.choice([True, False]):
                clause.append(var)
            else:
                clause.append(-var)
        # Clause completed. Appended it to the formula
        formula.append(clause)
    # Done
    return formula


def literal_value(literal: int, assignment: list[bool]) -> bool:
    """
    Return the truth value of a literal under a given assignment.

    assignment[i] corresponds to variable x_(i+1).
    """
    var_index = abs(literal) - 1
    value = assignment[var_index]

    if literal > 0:
        return value
    return not value


def clause_value(clause: list[int], assignment: list[bool]) -> bool:
    """
    Return True if at least one literal in the clause is True.
    """
    for literal in clause:
        if literal_value(literal, assignment):
            return True
    return False


def formula_value(formula: list[list[int]], assignment: list[bool]) -> bool:
    """
    Return True if all clauses are satisfied.
    """
    for clause in formula:
        if not clause_value(clause, assignment):
            return False
    return True


def number_to_assignment(number: int, num_vars: int) -> list[bool]:
    """
    Convert an integer to a truth assignment.

    Bit i of number becomes truth value of variable x_(i+1).
    """
    assignment: list[bool] = []

    for i in range(num_vars):
        bit = (number >> i) & 1
        assignment.append(bit == 1)
    print(assignment)
    return assignment


def brute_force_3sat(formula: list[list[int]], num_vars: int) -> tuple[bool, list[bool] | None, int]:
    """
    Try every possible assignment.

    Returns:
        (is_satisfiable, satisfying_assignment_or_None, assignments_checked)
    """
    total_assignments = 2 ** num_vars

    for number in range(total_assignments):
        assignment = number_to_assignment(number, num_vars)
        if formula_value(formula, assignment):
            return True, assignment, number + 1

    return False, None, total_assignments


def format_assignment(assignment: list[bool]) -> str:
    """
    Pretty-print an assignment.
    """
    parts: list[str] = []

    for i, value in enumerate(assignment):
        parts.append(f"x{i + 1}={value}")

    return ", ".join(parts)


def run_demo() -> None:
    """
    Run timing experiments for increasing numbers of variables.
    """
    random.seed(42)

    print("3SAT brute-force demonstration")
    print("-" * 50)
    print("n = number of variables")
    print("We check up to 2^n assignments.")
    print()

    for num_vars in range(4, 6):
        num_clauses = 4 * num_vars
        formula = make_random_3sat(num_vars, num_clauses)

        start = time.perf_counter()
        satisfiable, assignment, checked = brute_force_3sat(formula, num_vars)
        end = time.perf_counter()

        elapsed = end - start
        theoretical = 2 ** num_vars

        print(
            f"n={num_vars:2d} | "
            f"clauses={num_clauses:3d} | "
            f"max assignments={theoretical:10d} | "
            f"checked={checked:10d} | "
            f"time={elapsed:8.4f} sec | "
            f"SAT={satisfiable}"
        )

        # Stop if runtime is getting too large for a classroom demo
        if elapsed > 5.0:
            print()
            print("Stopping early because runtime is already getting large.")
            break


if __name__ == "__main__":
    run_demo()
```
